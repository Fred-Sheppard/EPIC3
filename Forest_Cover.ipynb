# Pre Processing for Forest Cover Data

## Explanation
 We have forest cover data only for specific years, and the objective is to use interpolation to obtain a value for each individual year from 1990 to 2019 instead of our current state where we only have certain years.

```python
import numpy as np
import csv

# Pre Processing forest growth data to include all years
# Forest Growth Data
years = np.array([ 1950, 1965, 1973, 1985, 2006, 2012, 2017, 2022])
values = np.array([ 1.40, 3.70, 4.60, 5.90, 10.10, 10.50, 11.00, 11.60]) # Percentage of Irish land covered in forest
```

# Data before pre-processing

Missing values for certain years, as shown on graph below


```python
import matplotlib.pyplot as plt

plt.plot(years, values, 'o--')
plt.xlabel('Year')
plt.ylabel('Percentage Forest Cover')
plt.title('Percentage Forest Cover over Years')
plt.show()
```

```python
# Create an array of years from the first year to the last year
all_years = np.arange(years[0], years[-1] + 1)

# Interpolate values for all years - finds value for each year from 1950-2022 
interpolated_values = np.interp(all_years, years, values)

# Write to CSV file
with open('Forest_Cover_All_Years.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Year', 'Value'])
    for year, value in zip(all_years, interpolated_values): # Couple each single year with each single value using zip
        if 1990 <= year <= 2019 :
            writer.writerow([year, value])
```

# Data after using interpolation to generate missing years

shown on graph below

```python

plt.plot(all_years, interpolated_values, 'o--')
#plt.plot(years, values, 'o--')
plt.xlabel('Year')
plt.ylabel('Percentage Forest Cover')
plt.title('Percentage Forest Cover over Years')
plt.show()
```
